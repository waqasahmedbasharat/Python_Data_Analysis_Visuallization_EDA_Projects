## Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

## Dropping irrelevant columns

```python
df = df.drop(['Vehicle Size'], axis=1)
df.head()
```

## Renaming the columns

```python
df = df.rename(columns={'Engine Fuel Type': 'Fuel', 'Engine HP':'HP', 'Engine Cylinders': 'Cylinders', 'MSRP':'Price'})
df.head()
```

## Finding duplicate rows

```python
df.shape
```

```python
duplicate_rows = df[df.duplicated()]
print("Duplicate rows are: ", duplicate_rows.shape)
```

## Counting rows in each column

```python
df.count()
```

## Dropping duplicate rows

```python
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

## Sum of null values

```python
print(df.isnull().sum())
```

## Dropping the missing values

```python
df = df.dropna()
df.count()
```

```python
#after dropping the missing values
print(df.isnull().sum())
```

## Detecting Outliers

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

## Manually finding IQR

```python
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)

IQR = Q3 - Q1
print(IQR)
```

## Seaborn on Titanic Dataset

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.shape
```

## 1. Distribution Plots

### hist plot

```python
sns.histplot(df['Parch'], kde=False)
plt.show()
```

### Kde Plot

```python
sns.distplot(df['Age'], hist=False)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.distplot(df['Age'])
plt.show()
```

## 2. Relational Plots

### Rel plot

```python
sns.relplot(x="Age", y="Fare", col="Pclass", hue="Sex", style="Sex", kind="line", data=df)
plt.show()
```

### Scatter plot

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x="Age", y="Fare", hue="Sex", data=df)
plt.show()
```

### line plot

```python
plt.figure(figsize=(8,8))
sns.lineplot(x="Age", y="Fare", hue="Sex", style="Sex", data=df)
plt.show()
```

## 3. Categorical Plots

### Bar plots

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=df)
plt.show()
```

### strip plot

```python
plt.figure(figsize=(8,8))
sns.stripplot(x="Sex", y="Age", hue="Sex", data=df)
plt.show()
```

### swarm plot

```python
plt.figure(figsize=(8,8))
sns.swarmplot(x="Sex", y="Age", hue="Sex", data=df)
plt.show()
```

### box plot

```python
plt.figure(figsize=(8,8))
sns.boxplot(x="Survived", y="Age", data=df)
plt.show()
```

### Violin plot

```python
plt.figure(figsize=(8,8))
sns.violinplot(x="Survived", y="Age", hue="Sex", data=df)
plt.show()
```

### count plot

```python
sns.countplot(x="Survived", data=df, palette="Blues")
plt.show()
```

### point plot

```python
sns.pointplot(x="Sex", y="Survived", hue="Pclass", data=df)
plt.show()
```

## 4. Regression Plots

### lmplot

```python
sns.lmplot(x="Age", y="Fare", data=df)
plt.show()
```

## reg plot

```python
sns.regplot(x="Age", y="Fare", data=df)
plt.show()
```

```python
# strong positive corr
sns.lmplot(x="Fare", y="Fare", data=df)
plt.show()
```

## 5. Matrix Plots

### Heatmap

```python
sns.heatmap(df.corr(), cmap="YlGnBu", annot=True, fmt = ".2f")
plt.show()
```

## 6. Multi plot grids

### Facet Grid Map

```python
g = sns.FacetGrid(df, col = "Survived", row="Pclass")

g.map(plt.hist, 'Age')
g.add_legend()
plt.show()
```
